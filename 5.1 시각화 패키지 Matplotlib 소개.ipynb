### 틱 설정
- 플롯이나 차트에서 축상의 위치 표시 지점을 틱(tick)이라고 하고 이 틱에 써진 숫자 혹은 글자를 틱 라벨(tick label)이라고 한다.

```python
x = np.linspace(-np.pi,np.pi,256) # (구간 시작점,구간 끝점,구간 내 숫자 개수)
c=np.cos(x)
plt.title('x축과 y축의 tick label 설정')
plt.plot(x,c)
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi])
plt.yticks([-1,0,+1])
plt.show()
```

- 틱 라벨 문자열에 $$ 사이에 LaTeX 수학 문자식을 넣기

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("LaTeX, 문자열로 tick label 정의")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.show()
```

### 그리드 설정

### 여러개의 선을 그리기
- 라인 플롯에서 선을 하나가 아니라 여러개를 그리고 싶은 경우
- x 데이터, y 데이터, 스타일 문자열을 반복하여 인수로 넘긴다.
- 하나의 선을 그릴 때 처럼 x 데이터나 스타일 문자열을 생략할 수 없다.

```python
t = np.arange(0.,5.,0.2)
plt.title('라인 플롯에서 여러개의 선 그리기')
plt.plot(t,t,'r--',t,0.5*t**2, 'bs:',t,0.2*t**3, 'g^-')
plt.show()
```

### 범례
- 범례의 위치는 자동으로 정해지지만 수동으로 설정하고 싶으면 loc 인수를 사용한다.

```python
x = np.linspace(-np.pi,np.pi,256)
c,s=np.cos(x),np.sin(x)
plt.title('legend를 표시한 플롯')
plt.plot(x,c,ls='--',label='cosine')
plt.plot(x,s,ls=':',label='sine')
plt.legend(loc=2) # upper right
plt.show()
```

### Axes 객체와 subplot 명령
- subplot 명령은 그리드 형태의 Axes 객체들을 생성하는데 Figure가 행렬이고 Axes가 행렬의 원소라고 생각하면 된다.

http://matplotlib.org/api/axes_api.html#matplotlib.axes.Axes

```python
x1 = np.linspace(0.0,5.0)
x2 = np.linspace(0.0,2.0)
y1 = np.cos(2*np.pi*x1)*np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

ax1 = plt.subplot(2,1,1)
plt.plot(x1,y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

ax2 = plt.subplot(2,1,2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

- 2x2 형태의 네 개의 플롯일 때
- subplot의 인수는 (2,2,1)를 줄여서 221 라는 하나의 숫자로 표시할 수도 있다.

```python
np.random.seed(0)

plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title('axes 1')

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title('axes 2')

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title('axes 3')

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title('axes 4')

plt.tight_layout()
plt.show()
```

```python
fig,axes = plt.subplots(2,2)

np.random.seed(0)
axes[0,0].plot(np.random.rand(5))
axes[0,0].set_title('axes 1')
axes[0,1].plot(np.random.rand(5))
axes[0,1].set_title('axes 2')
axes[1,0].plot(np.random.rand(5))
axes[1,0].set_title('axes 3')
axes[1,1].plot(np.random.rand(5))
axes[1,1].set_title('axes 4')

plt.tight_layout()
plt.show()
```

### Axis 객체와 축
- 여러가지 플롯을 하나의 Axes 객체에 표시할 때 y값의 크기가 달라서 표시하기 힘든 경우

https://matplotlib.org/api/axis_api.html

```python
fig,ax0 = plt.subplots()
ax1 = ax0.twinx()
ax0.set_title('2개의 y축 한 figure에서 사용하기')
ax0.plot([10,5,2,9,7],'r-',label='y0')
ax0.set_ylabel('y0')
ax0.grid(False)
ax1.plot([100,200,220,180,120],'g:',label='y1')
ax1.set_ylabel('y1')
ax1.grid(False)
ax0.set_xlabel('공유되는 x축')
plt.show()
```
